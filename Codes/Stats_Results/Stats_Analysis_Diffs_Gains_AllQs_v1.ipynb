### Statisical analysis on some results of the four models

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
current_dir = os.getcwd()

print("Current directory:", current_dir)

# Name the new folder
Output_folder_main='Stats_Analysis_Diffs_Gains_AllQs_v1'
```

```python
# Output_folder = Output_folder_main+'/SampLSTM_'+str(SampNo_lstm)+'/NMS_'+str(nth_mid_sth_fg)+'/EffTp_'+str(Eff_limit_combs[0])+'/GrOck_'+str(gr_ock_fg)+'_loc_'+str(local_Eff_fg)+'_inps_'+str(var_inputs_combs[0])+'_lag_'+str(lag_test_combs[0])

Output_folder=Output_folder_main
## Create the output folder
folder_path = r''+Output_folder
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
df_meas_pred_q={}

for Qr in ['Top75', 'Top50', 'Top25', 'Top10']:
    df_list = []
    for mod in ['LSTM', 'DTW', 'KNN', 'HMM']:
        file_name = mod + '_Eff_test_meas_pred_' + Qr + '.xlsx'
        df_mod_q = pd.read_excel('Results/' + file_name)
        df_mod_q = df_mod_q.rename(columns={'Unnamed: 0': 'index'})
        df_meas = df_mod_q[['index', 'Voyage', 'Weather_State', 'Weather_String', 'fuel_meas', 'time_meas', 'eff_meas']]
        
        # Rename columns with suffix based on 'mod'
        cols_old = ['fuel_pred', 'time_pred', 'eff_pred', 'eff_diff', 'eff_gain']
        cols_new = ['fuel_pred_'+mod, 'time_pred_'+mod, 'eff_pred_'+mod, 'eff_diff_'+mod, 'eff_gain_'+mod]
        column_mapping = dict(zip(cols_old, cols_new))
        df_mod_q = df_mod_q.rename(columns=column_mapping)

        # Append the concatenated DataFrame to the list
        df_list.append(df_mod_q[cols_new])


    # Create the final DataFrame by concatenating all DataFrames in the list side by side
    df_meas_pred_q[Qr] = pd.concat(df_list, axis=1)
    df_meas_pred_q[Qr] = pd.concat([df_meas, df_meas_pred_q[Qr]], axis=1)

# df_meas_pred_q
```

```python
df_meas_pred_q['Top10']
```

```python
##import data
#############
df_90 = df_meas_pred_q['Top10']
df_75 = df_meas_pred_q['Top25']
df_50 = df_meas_pred_q['Top50']
df_25 = df_meas_pred_q['Top75']


df_90 = df_90.drop(df_90.index[:-2])
df_90.index = ['Top10', 'Top10']
df_75 = df_75.drop(df_75.index[:-2])
df_75.index = ['Top25', 'Top25']
df_50 = df_50.drop(df_50.index[:-2])
df_50.index = ['Top50', 'Top50']
df_25 = df_25.drop(df_25.index[:-2])
df_25.index = ['Top75', 'Top75']
```

```python
df_stats= pd.concat([df_25, df_50, df_75, df_90], axis=0)

print(df_stats.shape)
print(648/4)
```

```python
df_stats
```

### Statistics of Eff Gains 

```python
filtered_columns = df_stats.filter(like='eff', axis=1)
filtered_columns = filtered_columns.drop(columns=filtered_columns.filter(like='eff_meas').columns)
filtered_columns = filtered_columns.drop(columns=filtered_columns.filter(like='eff_pred').columns)
filtered_columns = filtered_columns.drop(columns=filtered_columns.filter(like='eff_diff').columns)

filtered_columns.insert(0, 'index', df_stats['index'])
filtered_columns
```

```python
df_stats_sum = filtered_columns[df_stats['index'] != 'avg']
df_stats_avg = filtered_columns[df_stats['index'] != 'sum']

# Calculate the average of all columns
average_row = df_stats_sum.select_dtypes(include='number').mean()
df_stats_sum = df_stats_sum.append(average_row, ignore_index=True)
df_stats_sum = df_stats_sum.rename(index={df_stats_sum.index[-1]: 'average'})

# Calculate the average of all columns
average_row = df_stats_avg.select_dtypes(include='number').mean()
df_stats_avg = df_stats_avg.append(average_row, ignore_index=True)
df_stats_avg = df_stats_avg.rename(index={df_stats_avg.index[-1]: 'average'})

df_stats_sum.index = ['Top75', 'Top50', 'Top25', 'Top10', 'average']
df_stats_avg.index = ['Top75', 'Top50', 'Top25', 'Top10', 'average']

# Remove 'eff_gain' from the column names
df_stats_avg.columns = df_stats_avg.columns.str.replace('eff_gain_', '')
df_stats_avg = df_stats_avg.iloc[:, 1:]

# Remove 'eff_gain' from the column names
df_stats_sum.columns = df_stats_sum.columns.str.replace('eff_gain_', '')
df_stats_sum = df_stats_sum.iloc[:, 1:]


# Save the results to an Excel file
df_stats_sum.round(2).to_excel(Output_folder+"/All_eff_gains_sum.xlsx")
# Save the results to an Excel file
df_stats_avg.round(2).to_excel(Output_folder+"/All_eff_gains_avg.xlsx")

df_stats_avg
```
