### Statisical analysis on some results of the four models

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
current_dir = os.getcwd()

print("Current directory:", current_dir)

# Name the new folder
Output_folder_main='Stats_Analysis_Counts_AllQs_v1'
```

```python
# Output_folder = Output_folder_main+'/SampLSTM_'+str(SampNo_lstm)+'/NMS_'+str(nth_mid_sth_fg)+'/EffTp_'+str(Eff_limit_combs[0])+'/GrOck_'+str(gr_ock_fg)+'_loc_'+str(local_Eff_fg)+'_inps_'+str(var_inputs_combs[0])+'_lag_'+str(lag_test_combs[0])

Output_folder=Output_folder_main
## Create the output folder
folder_path = r''+Output_folder
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
df_meas_pred_q={}

for Qr in ['Top75', 'Top50', 'Top25', 'Top10']:
    df_list = []
    for mod in ['LSTM', 'DTW', 'KNN', 'HMM']:
        file_name = mod + 'Counts_Perc_Eff_Test_' + Qr + '.xlsx'
        df_mod_q = pd.read_excel('Results/' + file_name)
        df_mod_q = df_mod_q.rename(columns={'Unnamed: 0': 'index'})
        df_meas = df_mod_q['index']
        
        # Rename columns with suffix based on 'mod'
        cols_old = ['Improved_Fuel_Vs', 'Not_Improved_Fuel_Vs', 'Improved_Time_Vs', 'Not_Improved_Time_Vs', 
                    'Improved_Eff_Vs', 'Not_Improved_Eff_Vs']
        cols_new = ['Improved_Fuel_Vs'+mod, 'Not_Improved_Fuel_Vs'+mod, 'Improved_Time_Vs'+mod, 'Not_Improved_Time_Vs'+mod,
                    'Improved_Eff_Vs'+mod, 'Not_Improved_Eff_Vs'+mod]
        column_mapping = dict(zip(cols_old, cols_new))
        df_mod_q = df_mod_q.rename(columns=column_mapping)

        # Append the concatenated DataFrame to the list
        df_list.append(df_mod_q[cols_new])


    # Create the final DataFrame by concatenating all DataFrames in the list side by side
    df_m_p_q = pd.concat(df_list, axis=1)
    df_m_p_q = pd.concat([df_meas, df_m_p_q], axis=1)
    
    df_m_p_q.index = [Qr, Qr]
    df_meas_pred_q[Qr]=df_m_p_q


# df_meas_pred_q
```

```python
##import data
#############
df_90 = df_meas_pred_q['Top10']
df_75 = df_meas_pred_q['Top25']
df_50 = df_meas_pred_q['Top50']
df_25 = df_meas_pred_q['Top75']
```

```python
df_stats= pd.concat([df_25, df_50, df_75, df_90], axis=0)


# Save the results to an Excel file
df_stats.to_excel(Output_folder+"/All_Counts_Prec.xlsx")


print(df_stats.shape)
print(648/4)
df_stats
```

```python
df_stats_count = df_stats[df_stats['index'] != 'Percentage']
df_stats_perc = df_stats[df_stats['index'] != 'Count']

# Calculate the average of all columns
average_row = df_stats_count.select_dtypes(include='number').mean()
df_stats_count = df_stats_count.append(average_row, ignore_index=True)
df_stats_count = df_stats_count.rename(index={df_stats_count.index[-1]: 'average'})

# Calculate the average of all columns
average_row = df_stats_perc.select_dtypes(include='number').mean()
df_stats_perc = df_stats_perc.append(average_row, ignore_index=True)
df_stats_perc = df_stats_perc.rename(index={df_stats_perc.index[-1]: 'average'})

df_stats_count.index = ['Top75', 'Top50', 'Top25', 'Top10', 'average']
df_stats_perc.index = ['Top75', 'Top50', 'Top25', 'Top10', 'average']


# Save the results to an Excel file
df_stats_count.to_excel(Output_folder+"/All_Counts.xlsx")
# Save the results to an Excel file
df_stats_perc.to_excel(Output_folder+"/All_Prec.xlsx")

df_stats_perc
```
