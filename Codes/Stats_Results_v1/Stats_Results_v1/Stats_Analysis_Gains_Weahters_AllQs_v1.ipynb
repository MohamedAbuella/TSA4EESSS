```python
df_meas_pred_q={}

for Qr in ['Top75', 'Top50', 'Top25', 'Top10']:
    df_list = []
    for mod in ['LSTM', 'DTW', 'KNN', 'HMM']:
        file_name = mod + '_Eff_test_meas_pred_' + Qr + '.xlsx'
        df_mod_q = pd.read_excel('Results/' + file_name)
        df_mod_q = df_mod_q.rename(columns={'Unnamed: 0': 'index'})
        df_meas = df_mod_q[['index', 'Voyage', 'Weather_State', 'Weather_String', 'fuel_meas', 'time_meas', 'eff_meas']]
        
        # Rename columns with suffix based on 'mod'
        cols_old = ['fuel_pred', 'time_pred', 'eff_pred', 'eff_diff', 'eff_gain']
        cols_new = ['fuel_pred_'+mod, 'time_pred_'+mod, 'eff_pred_'+mod, 'eff_diff_'+mod, 'eff_gain_'+mod]
        column_mapping = dict(zip(cols_old, cols_new))
        df_mod_q = df_mod_q.rename(columns=column_mapping)

        # Append the concatenated DataFrame to the list
        df_list.append(df_mod_q[cols_new])


    # Create the final DataFrame by concatenating all DataFrames in the list side by side
    df_meas_pred_q[Qr] = pd.concat(df_list, axis=1)
    df_meas_pred_q[Qr] = pd.concat([df_meas, df_meas_pred_q[Qr]], axis=1)

# df_meas_pred_q
```

```python
##import data
#############
df_90 = df_meas_pred_q['Top10']
df_75 = df_meas_pred_q['Top25']
df_50 = df_meas_pred_q['Top50']
df_25 = df_meas_pred_q['Top75']

df_90.fillna(0, inplace=True)
df_90 = df_90.drop(df_90.index[-2:])
df_75.fillna(0, inplace=True)
df_75 = df_75.drop(df_75.index[-2:])
df_50.fillna(0, inplace=True)
df_50 = df_50.drop(df_50.index[-2:])
df_25.fillna(0, inplace=True)
df_25 = df_25.drop(df_25.index[-2:])
```

```python
df_stats= pd.concat([df_25, df_50, df_75, df_90], axis=0)

print(df_stats.shape)
print(648/4)
```

```python
# Assuming 'df_stats' is your DataFrame

# Select only the relevant columns for Eff_diff in each model
eff_diff_columns = ['eff_diff_LSTM', 'eff_diff_DTW', 'eff_diff_KNN', 'eff_diff_HMM']

# Calculate the standard deviation for Eff_diff in each model
df_stats['std_dev_eff_diff'] = df_stats[eff_diff_columns].std(axis=1)

# Find the row with the highest standard deviation for Eff_diff
row_with_highest_std_dev = df_stats.loc[df_stats['std_dev_eff_diff'].idxmax()]

# Print the row with the highest standard deviation
print("Row with the highest standard deviation in Eff_diff:", row_with_highest_std_dev)
```

```python
# Extract unique weather conditions
weather_conditions = df_stats['Weather_String'].unique()

# Calculate average efficiency improvement for each weather condition
average_efficiency_improvement = {}
for weather_condition in weather_conditions:
    subset = df_stats[df_stats['Weather_String'] == weather_condition]
    average_efficiency_improvement[weather_condition] = subset['eff_gain_HMM'].mean()  # Replace 'eff_gain_HMM' with the desired model

# Print the results
for weather, avg_improvement in average_efficiency_improvement.items():
    print(f"{weather}: {avg_improvement:.2f}%")

# Find the weather condition with the highest average efficiency improvement
max_condition = max(average_efficiency_improvement, key=average_efficiency_improvement.get)
print(f"{max_condition} has the highest average efficiency improvement.")
```

```python
# Extract unique weather conditions
weather_conditions = df_stats['Weather_String'].unique()

# Create a dictionary to store average efficiency improvements for each model
average_efficiency_improvements = {}

# List of models to consider
models = ['eff_gain_LSTM', 'eff_gain_DTW', 'eff_gain_KNN', 'eff_gain_HMM']

# Calculate average efficiency improvement for each model and weather condition
for model in models:
    average_efficiency_improvement = {}
    for weather_condition in weather_conditions:
        subset = df_stats[df_stats['Weather_String'] == weather_condition]
        average_efficiency_improvement[weather_condition] = subset[model].mean()

    average_efficiency_improvements[model] = average_efficiency_improvement

# Create a DataFrame from the results
result_df = pd.DataFrame(average_efficiency_improvements).round(2)

desired_order = ["calm_weather", "moderate_weather", "extreme_weather", "Average"]
result_df = result_df.reindex(desired_order)

# Add the averages as a new row with the label "Average"
result_df.loc['Average'] = round(result_df.mean(),2)

# Save the results to an Excel file
result_df.to_excel(Output_folder+"/All_Avg_eff_gains_weathers.xlsx", index_label="Weather Condition")

result_df
```

```python
# Create a dictionary to store standard deviations of efficiency improvements for each model
std_deviation_efficiency_improvements = {}

# Calculate standard deviation of efficiency improvement for each model and weather condition
for model in models:
    std_deviation_efficiency_improvement = {}
    for weather_condition in weather_conditions:
        subset = df_stats[df_stats['Weather_String'] == weather_condition]
        std_deviation_efficiency_improvement[weather_condition] = subset[model].std()

    std_deviation_efficiency_improvements[model] = std_deviation_efficiency_improvement

# Create a DataFrame from the results
std_df = pd.DataFrame(std_deviation_efficiency_improvements).round(2)

desired_order = ["calm_weather", "moderate_weather", "extreme_weather", "Average"]
std_df = std_df.reindex(desired_order)

# Add the averages as a new row with the label "Average"
std_df.loc['Average'] = round(std_df.mean(),2)

# Save the standard deviation results to an Excel file
std_df.to_excel(Output_folder + "/All_Std_eff_gains_weathers.xlsx", index_label="Weather Condition")

std_df
```

```python
import pandas as pd

# 1. Average Efficiency Improvement for Different Models
average_efficiency_improvement_by_model = {}
models = ['eff_gain_LSTM', 'eff_gain_DTW', 'eff_gain_KNN', 'eff_gain_HMM']
for model in models:
    average_efficiency_improvement_by_model[model] = df_stats[model].mean()

# 2. Identify the Model with the Highest Efficiency Improvement
best_model = max(average_efficiency_improvement_by_model, key=average_efficiency_improvement_by_model.get)

# 3. Statistical Tests for Model Comparison
# Perform statistical tests to compare the efficiency improvements among different models.

# 4. Weather Condition Analysis for Each Model
average_efficiency_improvement_by_weather = {}
for model in models:
    for weather_condition in weather_conditions:
        subset = df_stats[df_stats['Weather_String'] == weather_condition]
        average_efficiency_improvement_by_weather[(model, weather_condition)] = subset[model].mean()

# 5. Identify Weather Conditions with the Most Variability in Efficiency Improvement
variability_by_weather = {}
for model in models:
    for weather_condition in weather_conditions:
        subset = df_stats[df_stats['Weather_String'] == weather_condition]
        variability_by_weather[(model, weather_condition)] = subset[model].std()
        
        
# 6. Top Performing Weather Conditions for Each Model
top_performing_conditions_by_model = {}
for model in models:
    top_performing_condition = max(average_efficiency_improvement_by_weather, key=lambda k: average_efficiency_improvement_by_weather[k] if k[0] == model else 0)
    top_performing_conditions_by_model[model] = {
        'Weather Condition': top_performing_condition[1],
        'Efficiency Improvement': average_efficiency_improvement_by_weather[top_performing_condition]
    }

# Create DataFrames for the results
df_stats_avg_improvement_by_model = pd.DataFrame(average_efficiency_improvement_by_model, index=['Value']).T
df_stats_avg_improvement_by_weather = pd.DataFrame.from_dict(average_efficiency_improvement_by_weather, orient='index').reset_index()
df_stats_variability_by_weather = pd.DataFrame.from_dict(variability_by_weather, orient='index').reset_index()
df_stats_top_conditions_by_model = pd.DataFrame(top_performing_conditions_by_model).T

# # Save the DataFrames to Excel files
# with pd.ExcelWriter('analysis_results.xlsx') as writer:
#     df_stats_avg_improvement_by_model.to_excel(writer, sheet_name='Average_Improvement_by_Model')
#     df_stats_avg_improvement_by_weather.to_excel(writer, sheet_name='Average_Improvement_by_Weather')
#     df_stats_variability_by_weather.to_excel(writer, sheet_name='Variability_by_Weather')
#     df_stats_top_conditions_by_model.to_excel(writer, sheet_name='Top_Conditions_by_Model')
```
